```python
import findspark
findspark.init()

import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Data 1116").getOrCreate()
spark
```

```python
path = "C:\Motorola\practice data (raw)\Data 1116/"
```

```python
df = spark.read.csv(path+'part*',header = False,inferSchema=True)
```

```python
df.toPandas()
```

```python
df.printSchema()
```

```python
headers = ["agmbgm","business_unit","dateTime","chargetype","cluster_cost","cost","cost_location","currency","departmentCode","environment","exec_owner","finance_owner","friendly_name","is_cie_project","meter","meter_subcategory","pricingmodel","product","product_owner","project_name","res_name","resource_group","run_date","service_name","service_tier","sleeve","unique_id","to_date","to_hour"]
```

```python
df = df.toDF(*headers)
df.toPandas()
```

```python
df.printSchema()
```

# Count Total Data

```python
df.count()
```

# Removing Duplicate

```python
df_dsc = df.orderBy("unique_id","dateTime",ascending = False)
```

```python
df_dsc.toPandas()
```

req non repeating with latest date

```python
df_asc = df_dsc.dropDuplicates(["unique_ID"]).orderBy("unique_ID")
df_asc.toPandas()
```

# Non repeating count

```python
df_asc.write.csv("Data 16-11")
```

```python
df_asc.count()
```

# Number of Duplicates

```python
a = df.count()
b = df_asc.count()
print(a - b)
```

# Cost per Environment

```python
df_asc.filter(df_asc.cluster_cost==True).groupBy('Environment').sum("cost").orderBy("Environment").toPandas()
```

# Total Cluster Cost

```python
df_asc.groupBy('cluster_cost').sum('cost').toPandas()
```
