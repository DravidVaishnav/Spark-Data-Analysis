```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
schm = StructType([
    StructField("agmbgm",StringType(),True),
    StructField("business_unit",StringType(),True),
    StructField("dateTime",TimestampType(),True),
    StructField("chargetype",StringType(),True),
    StructField("cluster_cost",BooleanType(),True),
    StructField("cost",DoubleType(),True),
    StructField("cost_location",StringType(),True),
    StructField("currency",StringType(),True),
    StructField("departmentCode",StringType(),True),
    StructField("environment",StringType(),True),
    StructField("exec_owner",StringType(),True),
    StructField("finance_owner",StringType(),True),
    StructField("friendly_name",StringType(),True),
    StructField("is_cie_project",BooleanType(),True),
    StructField("meter",StringType(),True),
    StructField("meter_subcategory",StringType(),True),
    StructField("pricingmodel",StringType(),True),
    StructField("product",StringType(),True),
    StructField("product_owner",StringType(),True),
    StructField("project_name",StringType(),True),
    StructField("res_name",StringType(),True),
    StructField("resource_group",StringType(),True),
    StructField("run_date",StringType(),True),
    StructField("service_name",StringType(),True),
    StructField("service_tier",StringType(),True),
    StructField("sleeve",StringType(),True),
    StructField("unique_id",LongType(),True),
    StructField("to_date",IntegerType(),True),
    StructField("to_hour",IntegerType(),True)
    
    
])
```

reading non rep data saved in previous task  

```python
df = spark.read.csv(path+'part*',schema = schm)
```

```python
df.toPandas()
```

new dataframe for env and total cost

```python
df_cost = df.filter(df.cluster_cost==True).groupBy("environment").agg(sum('cost').alias('total_cost')).orderBy('environment')
df_cost.toPandas()
```

joining both df

```python
df_req = df.filter(df.cluster_cost == True).join(df_cost,['environment'],"left")
```

```python
df_req.toPandas()
```

req percentage cost of project name per environment

```python
df_per = df_req.withColumn("cost_percent",df_req.cost/df_req.total_cost*100)
```

```python
df_per.toPandas()
```

This should return the cost percentage per environment for different projects but as here there is only 'None' project_name hence we get percentage cost 100 for every env

```python
df_per.groupBy("environment","project_name").agg(sum('cost_percent').alias('percentage_cost_env')).orderBy("environment").toPandas()
```

# **********Extra***************

We can apply same logic for service_tier instead of project_name and we can see results as there are many service_tiers in different environments

list of unique service_tiers

```python
df_per.dropDuplicates(['service_tier']).select("service_tier").orderBy("service_tier").toPandas()
```

percentage cost for service tier per environment

```python
df_per.groupBy("environment","service_tier").agg(sum('cost_percent').alias("per_cost")).orderBy("environment").toPandas()
```

if want to check highest costing service tiers

```python
df_per.groupBy("environment","service_tier").agg(sum('cost_percent').alias("per_cost")).orderBy("per_cost",ascending = False).toPandas()
```

if want to check for any particular environment

```python
df_per.filter(df_per.environment == "Australia Prod AustraliaCentral").groupBy("environment","service_tier").agg(sum('cost_percent').alias("per_cost")).orderBy("per_cost",ascending = False).toPandas()
```
